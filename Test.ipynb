```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
#plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.dpi'] = 140
```

```python
def S(x, a=1, h=0, v=0):
    return (1 / (1 + np.exp((-x*a) + h))) + v
```

```python
x = np.arange(-30, 60)
plt.plot(x, S(np.arange(-30, 60), a=0.2, h=0, v=0))
plt.plot(x, S(np.arange(-30, 60), a=1, h=35, v=0.4))
```
